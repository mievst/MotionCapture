```python
import torch
import torchvision
import cv2
import numpy as np
from torchvision.models.detection.keypoint_rcnn import keypointrcnn_resnet50_fpn, KeypointRCNN_ResNet50_FPN_Weights
```

```python
# Загрузка модели Keypoint RCNN из torchvision
model = torchvision.models.detection.keypoint_rcnn.keypointrcnn_resnet50_fpn(weights=KeypointRCNN_ResNet50_FPN_Weights)

# Перевод модели в режим оценки
model.eval()
```

```python
# Загрузка видео
cap = cv2.VideoCapture('dance.mp4')

while cap.isOpened():
    # Захват кадра
    ret, frame = cap.read()
    if not ret:
        break

    # Преобразование кадра в формат, необходимый для OpenPose
    input_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    image = torch.from_numpy(np.transpose(input_image, (2, 0, 1))).float()

    # Оценка позы объектов на изображении
    output = model([image])

    # Отображение кадра
    cv2.imshow('frame', frame)
    cv2.imshow('output', output)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Освобождение ресурсов
cap.release()
cv2.destroyAllWindows()
```
